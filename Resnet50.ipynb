```python
def DataSet():
    #训练集
    train_path_cane='raw-img/cane/'
    train_path_cavallo='raw-img/cavallo/'
    train_path_elefante='raw-img/elefante/'
    train_path_farfalla='raw-img/farfalla/'
    train_path_gallina='raw-img/gallina/'
    train_path_gatto='raw-img/gatto/'
    
    #测试集
    test_path_cane='txt/cane/'
    test_path_cavallo='txt/cavallo/'
    test_path_elefante='txt/elefante/'
    test_path_farfalla='txt/farfalla/'
    test_path_gallina='txt/gallina/'
    test_path_gatto='txt/gatto/'
    
    #罗列出训练集的数据
    imglist_train_cane=os.listdir(train_path_cane)
    imglist_train_cavallo=os.listdir(train_path_cavallo)
    imglist_train_elefante=os.listdir(train_path_elefante)
    imglist_train_farfalla=os.listdir(train_path_farfalla)
    imglist_train_gallina=os.listdir(train_path_gallina)
    imglist_train_gatto=os.listdir(train_path_gatto)
    
    
    #罗列出测试集的数据
    imglist_test_cane=os.listdir(test_path_cane)
    imglist_test_cavallo=os.listdir(test_path_cavallo)
    imglist_test_elefante=os.listdir(test_path_elefante)
    imglist_test_farfalla=os.listdir(test_path_farfalla)
    imglist_test_gallina=os.listdir(test_path_gallina)
    imglist_test_gatto=os.listdir(test_path_gatto)
    
    #处理训练集图片
    X_train = np.empty((len(imglist_train_cane) + len(imglist_train_cavallo)+ len(imglist_train_elefante)+len(imglist_train_farfalla)+len(imglist_train_gallina)+len(imglist_train_gatto), 224, 224, 3))
    Y_train = np.empty((len(imglist_train_cane) + len(imglist_train_cavallo)+ len(imglist_train_elefante)+len(imglist_train_farfalla)+len(imglist_train_gallina)+len(imglist_train_gatto), 6))
    count=0                   
    for img_name in imglist_train_cane:
                       img_path = train_path_cane + img_name
                       img=image.load_img(img_path, target_size=(224, 224))
                       img=image.img_to_array(img)/255.0
                       X_train[count] = img
                       Y_train[count] = np.array((1,0,0,0,0,0))
                       count+=1
    for img_name in imglist_train_cavallo:
                       img_path = train_path_cavallo + img_name
                       img=image.load_img(img_path, target_size=(224, 224))
                       img=image.img_to_array(img)/255.0
                       X_train[count] = img
                       Y_train[count] = np.array((0,1,0,0,0,0))
                       count+=1
    for img_name in imglist_train_elefante:
                       img_path = train_path_elefante + img_name
                       img=image.load_img(img_path, target_size=(224, 224))
                       img=image.img_to_array(img)/255.0
                       X_train[count] = img
                       Y_train[count] = np.array((0,0,1,0,0,0))
                       count+=1
    for img_name in imglist_train_farfalla:
                       img_path = train_path_farfalla + img_name
                       img=image.load_img(img_path, target_size=(224, 224))
                       img=image.img_to_array(img)/255.0
                       X_train[count] = img
                       Y_train[count] = np.array((0,0,0,1,0,0))
                       count+=1
    for img_name in imglist_train_gallina:
                       img_path = train_path_gallina + img_name
                       img=image.load_img(img_path, target_size=(224, 224))
                       img=image.img_to_array(img)/255.0
                       X_train[count] = img
                       Y_train[count] = np.array((0,0,0,0,1,0))
                       count+=1
    for img_name in imglist_train_gatto:
                       img_path = train_path_gatto + img_name
                       img=image.load_img(img_path, target_size=(224, 224))
                       img=image.img_to_array(img)/255.0
                       X_train[count] = img
                       Y_train[count] = np.array((0,0,0,0,0,1))
                       count+=1
    
    X_test = np.empty((len(imglist_test_cane) + len(imglist_test_cavallo)+ len(imglist_test_elefante)+len(imglist_test_farfalla) + len(imglist_test_gallina)+ len(imglist_test_gatto) , 224, 224, 3))
    Y_test = np.empty((len(imglist_test_cane) + len(imglist_test_cavallo)+ len(imglist_test_elefante)+len(imglist_test_farfalla) + len(imglist_test_gallina)+ len(imglist_test_gatto), 6))
    count=0
    for img_name in imglist_test_cane:
                       img_path = test_path_cane + img_name
                       img=image.load_img(img_path, target_size=(224, 224))
                       img=image.img_to_array(img)/255.0
                       X_test[count] = img
                       Y_test[count] = np.array((1,0,0,0,0,0))
                       count+=1
    for img_name in imglist_test_cavallo:
                       img_path = test_path_cavallo + img_name
                       img=image.load_img(img_path, target_size=(224, 224))
                       img=image.img_to_array(img)/255.0
                       X_test[count] = img
                       Y_test[count] = np.array((0,1,0,0,0,0))
                       count+=1
    for img_name in imglist_test_elefante:
                       img_path = test_path_elefante + img_name
                       img=image.load_img(img_path, target_size=(224, 224))
                       img=image.img_to_array(img)/255.0
                       X_test[count] = img
                       Y_test[count] = np.array((0,0,1,0,0,0))
                       count+=1
    for img_name in imglist_test_farfalla:
                       img_path = test_path_farfalla + img_name
                       img=image.load_img(img_path, target_size=(224, 224))
                       img=image.img_to_array(img)/255.0
                       X_test[count] = img
                       Y_test[count] = np.array((0,0,0,1,0,0))
                       count+=1
    for img_name in imglist_test_gallina:
                       img_path = test_path_gallina + img_name
                       img=image.load_img(img_path, target_size=(224, 224))
                       img=image.img_to_array(img)/255.0
                       X_test[count] = img
                       Y_test[count] = np.array((0,0,0,0,1,0))
                       count+=1
    for img_name in imglist_test_gatto:
                       img_path = test_path_gatto + img_name
                       img=image.load_img(img_path, target_size=(224, 224))
                       img=image.img_to_array(img)/255.0
                       X_test[count] = img
                       Y_test[count] = np.array((0,0,0,0,0,1))
                       count+=1
    #1图片处理                    
    train_path_cane = transforms.Compose(
    [transforms.RandomResizedCrop(size=256,scale=(1.8,1.0)),
    transforms.RandomRotation(degrees=15),
    transforms.RandomHorizontalFlip(),
    transforms.CenterCrop(size=224),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],
                         [0.229,0.224,0.225])
    ])
    #2图片处理
    train_path_cavallo = transforms.Compose(
    [transforms.RandomResizedCrop(size=256,scale=(1.8,1.0)),
    transforms.RandomRotation(degrees=15),
    transforms.RandomHorizontalFlip(),
    transforms.CenterCrop(size=224),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],
                         [0.229,0.224,0.225])
    ])
    #3图片处理    
    train_path_elefante = transforms.Compose(
    [transforms.RandomResizedCrop(size=256,scale=(1.8,1.0)),
    transforms.RandomRotation(degrees=15),
    transforms.RandomHorizontalFlip(),
    transforms.CenterCrop(size=224),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],
                         [0.229,0.224,0.225])
    ])
    #4图片处理    
    train_path_farfalla = transforms.Compose(
    [transforms.RandomResizedCrop(size=256,scale=(1.8,1.0)),
    transforms.RandomRotation(degrees=15),
    transforms.RandomHorizontalFlip(),
    transforms.CenterCrop(size=224),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],
                         [0.229,0.224,0.225])
    ])
    #5图片处理    
    train_path_gallina = transforms.Compose(
    [transforms.RandomResizedCrop(size=256,scale=(1.8,1.0)),
    transforms.RandomRotation(degrees=15),
    transforms.RandomHorizontalFlip(),
    transforms.CenterCrop(size=224),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],
                         [0.229,0.224,0.225])
    ])
    #6图片处理    
    train_path_gatto = transforms.Compose(
    [transforms.RandomResizedCrop(size=256,scale=(1.8,1.0)),
    transforms.RandomRotation(degrees=15),
    transforms.RandomHorizontalFlip(),
    transforms.CenterCrop(size=224),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],
                         [0.229,0.224,0.225])
    ])
        
    #打乱训练中的数据
    index = [i for i in range(len(X_train))]
    random.shuffle(index)
    X_train = X_train[index]                           
    Y_train = Y_train[index]
    #打乱测试中的数据
    index = [i for i in range(len(X_test))]
    random.shuffle(index)
    X_test = X_test[index]
    Y_test = Y_test[index]
    return X_train,Y_train,X_test,Y_test
                           
```

```python
X_train,Y_train,X_test,Y_test = DataSet()
```

```python
model.compile(optimizer=tf.train.AdamOptimizer(0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=1, batch_size=12)
```

```python
model.evaluate(X_test, Y_test, batch_size=6)
```
